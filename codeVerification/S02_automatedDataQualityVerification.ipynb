## Check for Duplicates

```python
from collections import Counter # Useful function
def checkForDuplicates(checklist):
    # duplicates = []
    if type(checklist) == list:
        print("Checklist is a list")
        
        if len(checklist) == len(set(checklist)):
            print("No duplicates found")
        else:
            Counter(checklist)
            print("Duplicates: ", [key for key in Counter(checklist).keys() if Counter(checklist)[key] > 1]) # Create a list based on the dictionary returned.
                    
    #print(duplicates)
    else:
        print("Error checklist is not a list")
```

```python
checkForDuplicates(financialRatios_benchmark)
```

```python
checkForDuplicates(financialRatios_comparison)
```

```python
# Detect Missing Financial Ratios Components (Once all mis-match Financial Ratio names have been corrected)
# Cross-check individual Financial Ratios with Financial Ratios present in the analysis with extra values (larger list)

# Check differences between financialRatios_benchmark and financialRatios_comparison, listing them on a separate data object
financialRatios_extra = set(financialRatios_benchmark).symmetric_difference(set(financialRatios_comparison))

# print(financialRatios_extra)
# print(len(financialRatios_extra))

if len(financialRatios_extra) == 0:
    print("No differences found.")
else:
    print("Differences detected.")
    print(financialRatios_extra)
```
