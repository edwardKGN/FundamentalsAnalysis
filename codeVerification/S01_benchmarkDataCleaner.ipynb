```python
import pandas as pd # Pre-requisite to run Pandas
from pandas import Series, DataFrame
import numpy as np # Pre-requisite to run numpy
```

```python
# Name of File
file_name = "Bursa_financialServices_benchmark"
sheet_name = "Sheet1"
```

```python
# Update to include Formatting
xlsx_file_name = file_name + ".xlsx"
```

```python
# Extract Key Data from Raw Benchmark Data
benchmark_raw = pd.read_excel(xlsx_file_name, sheet_name)
benchmark_raw
```

```python
benchmark_raw = benchmark_raw.dropna()
benchmark_raw
```

```python
benchmark_raw = benchmark_raw.reset_index(drop = True)
benchmark_raw
```

```python
output_file_name = file_name + "_cleaned" + ".xlsx"
benchmark_raw.to_excel(output_file_name)
```
