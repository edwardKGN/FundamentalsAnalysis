```python
import pandas as pd # Pre-requisite to run Pandas
from pandas import Series, DataFrame
import numpy as np # Pre-requisite to run numpy
```

```python
xlsx_name = "LPICapital_Insurance" # output file name
```

```python
# Read Benchmark (Hand-calculation Output)
benchmark = pd.read_excel("Bursa_financialServices_benchmark_cleaned.xlsx", "Sheet1", index_col = 0)
benchmark
```

```python
# Read Comparison (Code Output)
comparison =  pd.read_excel("FA_Bursa_FinancialServices.xlsx", "output", index_col = 0)
comparison
```

```python
deviation = pd.DataFrame({benchmark.columns[0]: benchmark[benchmark.columns[0]]}) # Initialize DataFrame
deviation
```

```python
for index, row in benchmark.iterrows():    
    benchmark_Values = ((benchmark.loc[benchmark[benchmark.columns[0]] == row.values[0]]).values[0])[1:] # Select only the values
    comparison_Values = ((comparison.loc[comparison[comparison.columns[0]] == row.values[0]]).values[0])[1:]
    
    #print("benchmark_Values: ", benchmark_Values)
    #print("comparison_Values: ", comparison_Values)
    
    if all(benchmark_Values):
        #print("Cond 1")
        deviation_Values = (benchmark_Values - comparison_Values)/benchmark_Values
    else:
        #print("Cond 2")
        deviation_Values = [0, 0, 0, 0, 0]
    
    #print(deviation_Values)
    
    for index, element in enumerate(list(benchmark.columns[1:len(benchmark.columns)])):
        deviation.loc[deviation[deviation.columns[0]] == row.values[0], element] = deviation_Values[index] # Insert at row and column deviation value

deviation
```

```python
deviation.to_excel(xlsx_name + "_deviations" + ".xlsx", sheet_name = 'output')
```
